```python
import matplotlib.pyplot as plt

# Assuming the text is stored in a file named 'data.txt'
filename = 'mask_values.txt'

# Read the file and store the mask scores and row numbers
row_numbers = []
mask_scores = []

with open(filename, 'r') as file:
    # Skip the header line
    next(file)
    
    for line in file:
        # Split the line by whitespace and extract the row number and mask score
        row_no, _, _, mask_score = line.split()
        
        # Append the extracted values to the respective lists
        row_numbers.append(int(row_no))
        mask_scores.append(float(mask_score))

# Plotting
plt.figure(figsize=(10,5))
plt.plot(row_numbers, mask_scores, marker='o')
plt.title('Mask Scores')
plt.xlabel('Row Number (No)')
plt.ylabel('Mask Score')
plt.grid(True)
plt.savefig('mask_score.png', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to round the number to the nearest 0.05
def round_to_nearest_05(number):
    return round(number * 20) / 20  # Multiply by 20, round it, then divide by 20 to get the nearest 0.05

# The text is stored in a file named 'mask_values.txt'
filename = 'mask_values.txt'

# Read the file and store the mask scores
mask_scores = []

with open(filename, 'r') as file:
    # Skip the header line
    next(file)
    
    for line in file:
        # Split the line by whitespace and extract the mask score
        _, _, _, mask_score = line.split()
        
        # Round the mask score to the nearest 0.05 and append to the list
        mask_scores.append(round_to_nearest_05(float(mask_score)))

# Define the bin edges for the histogram
bin_edges = np.arange(-0.025, 1.075, 0.05)  # Starts at -0.025 to include 0, and ends at 1.075 to include 1.05

# Plotting
plt.figure(figsize=(12,7))
plt.hist(mask_scores, bins=bin_edges, edgecolor='k', color='skyblue', alpha=0.7, rwidth=0.8, label='Mask Score Counts')

# Adding title and labels with larger fonts
plt.title('Distribution of Mask Scores', fontsize=16)
plt.xlabel('Mask Score', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Adding axis ticks
plt.xticks(np.arange(0, 1.1, 0.05), fontsize=12)
plt.yticks(fontsize=12)

# Adding a legend with larger font
plt.legend(fontsize=12)

# Adding a grid
plt.grid(axis='y', alpha=0.75)

# Saving the plot to a file with high resolution
plt.savefig('mask_values_dist.png', bbox_inches='tight', dpi=300)

# Displaying the plot
plt.show()
```
