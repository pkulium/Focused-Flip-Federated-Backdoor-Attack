```python
import matplotlib.pyplot as plt

# Assuming the text is stored in a file named 'data.txt'
filename = 'mask_values.txt'

# Read the file and store the mask scores and row numbers
row_numbers = []
mask_scores = []

with open(filename, 'r') as file:
    # Skip the header line
    next(file)
    
    for line in file:
        # Split the line by whitespace and extract the row number and mask score
        row_no, _, _, mask_score = line.split()
        
        # Append the extracted values to the respective lists
        row_numbers.append(int(row_no))
        mask_scores.append(float(mask_score))

# Plotting
plt.figure(figsize=(10,5))
plt.plot(row_numbers, mask_scores, marker='o')
plt.title('Mask Scores')
plt.xlabel('Row Number (No)')
plt.ylabel('Mask Score')
plt.grid(True)
plt.savefig('mask_score.png', bbox_inches='tight')
plt.show()
```

```python

import matplotlib.pyplot as plt
from collections import Counter

# Assuming the text is stored in a file named 'data.txt'
filename = 'mask_values.txt'

# Read the file and store the mask scores
mask_scores = []

with open(filename, 'r') as file:
    # Skip the header line
    next(file)
    
    for line in file:
        # Split the line by whitespace and extract the mask score
        _, _, _, mask_score = line.split()
        
        # Append the extracted value to the list
        if float(mask_score)) < 0.5:
            mask_scores.append(float(mask_score))

# Count the occurrences of each unique mask score
mask_score_counts = Counter(mask_scores)

# Extract unique mask scores and their corresponding counts
unique_mask_scores = list(mask_score_counts.keys())
counts = list(mask_score_counts.values())

# Plotting
plt.figure(figsize=(10,5))
plt.bar(unique_mask_scores, counts, color='blue', alpha=0.7)
plt.title('Mask Score Counts')
plt.xlabel('Mask Score')
plt.ylabel('Count')
plt.grid(True)
plt.savefig('mask_score_distribution.png', bbox_inches='tight')
plt.show()
```
